```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# Configuraciones de panda
pd.set_option('display.max_rows',500)
pd.set_option('display.max_columns',500)
pd.set_option('display.width', 1000)
```

```python
# Definir URL de la pagina
url = "https://www.sensacine.com/peliculas/taquilla/usa/week-2024-09-27/"
web = requests.get(url)
sp = BeautifulSoup(web.text)

# Bsucar clase por clase, por atributos y valor
movie_data = sp.find_all('tr',attrs={'class':'responsive-table-row'})

movie_title = []
movie_company = []
movie_total = []
movie_total_acumulado = []
movie_semana = []

# Extraer data de cada pelicula
for movie in movie_data:
    # Extraer titulos
    title = movie.find('a', attrs={'class','meta-title-link'})
    
    # Extraer Compañia
    company = movie.find('div', attrs={'class','meta-body'})
    
    # Extraer Total
    total = movie.find('td', attrs={'data-heading':'Total'})
    
    # Extraer total acumulado
    total_acumulado = movie.find('td', attrs={'data-heading':'Acumulativo'})
    
    # Extraer semana de cine
    semana = movie.find('td', attrs={'data-heading':'Semana'})
    
    # Recorrer toda la tabla
    if title:
        movie_title.append(title.text)
        
    if company:
        movie_company.append(company.text.strip())
        
    if total:
        clean_total = total.text.strip().replace("$", "").replace(" ", "")
        movie_total.append(float(clean_total))
    
    if total_acumulado:
        clean_total_acumulado = total_acumulado.text.strip().replace("$", "").replace(" ", "")
        movie_total_acumulado.append(float(clean_total_acumulado))
    
    if semana:
        movie_semana.append(int(semana.text.strip()))

df = pd.DataFrame({'Titulo':movie_title, 'Compañia':movie_company, 'Total':movie_total,'Total acumulado':movie_total_acumulado, 'Semana':movie_semana})
df.head(9)
# df.to_excel('C:/Users/ASUS/Documents/FEMT-COURSES/cine-2024-07-26.xlsx', index=False)
```
